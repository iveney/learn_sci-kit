```python
import numpy as np
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# mostly negative
y = np.random.random(1000) > 0.9 
# predict all as negative
y_pred = np.zeros((1000,1));
fpr, tpr, _ = roc_curve(y, y_pred)
plt.plot(fpr, tpr, 'g-')
plt.show()

# random prediction
y_pred = np.random.random(1000) > 0.5
fpr, tpr, _ = roc_curve(y, y_pred)
plt.plot(fpr, tpr, 'g-')
plt.show()
```

```python
print(auc(fpr, tpr))
```
