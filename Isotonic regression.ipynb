```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np

# make some data with noise
X, y = make_regression(n_samples=100, n_features=1, noise=30.0, bias=-5)
print(X.shape)
print(y.shape)
```

```python
# plot the data
plt.plot(X, y, 'r.')
plt.show()
```

```python
# linear regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X, y)
```

```python
# Isotonic regression
from sklearn.isotonic import IsotonicRegression
iso = IsotonicRegression()
iso.fit(X[:, 0], y)
```

```python
# plot a line of LR fit
x = np.arange(-2, 2, 0.1)
print(x.shape)
plt.plot(x, lr.predict(x[:, np.newaxis]), 'b.-')
plt.plot(x, iso.predict(x), 'g.-', markersize=12)
plt.plot(X, y, 'r.')
plt.show()
```
